Filter products with a price greater than $100


```python
expensive_products = df[df['Price'] > 100]
print(expensive_products)
```

Group by Category and calculate the average price


```python
average_price_by_category = df.groupby('Category')['Price'].mean()
print(average_price_by_category)
```

SORTING DATA ON PRICE

```python
sorted_df = df.sort_values(by='Price', ascending=False)
print(sorted_df)
```

ADDING A NEW COLUMN

```python
# Define a function to categorize price
def categorize_price(price):
    if price < 100:
        return 'Low'
    elif 100 <= price < 500:
        return 'Medium'
    else:
        return 'High'

# Apply the function to create a new column
df['Price_Category'] = df['Price'].apply(categorize_price)

# Display the updated DataFrame
print("DataFrame with Price_Category:")
print(df.head())
```
