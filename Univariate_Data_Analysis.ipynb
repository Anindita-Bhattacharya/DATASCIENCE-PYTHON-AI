Univariate Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset of students' test scores
data = {
    'Student': range(1, 11),
    'Test Score': [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Descriptive statistics
mean = df['Test Score'].mean()
median = df['Test Score'].median()
mode = df['Test Score'].mode()[0]  # mode returns a Series, take the first element
data_range = df['Test Score'].max() - df['Test Score'].min()
std_dev = df['Test Score'].std(ddof=0)  # Population standard deviation

# Print the results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode}")
print(f"Range: {data_range}")
print(f"Standard Deviation: {std_dev:.2f}")

# Visualization: Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['Test Score'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.legend()
plt.show()
```
