```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame with the provided statistics
data = {
    'Price': np.random.normal(loc=456.83, scale=398.17, size=8000).clip(5.12, 1499.72),
    'Customer_Age': np.random.normal(loc=41.61, scale=13.96, size=8000).clip(18, 65),
    'Purchase_History': np.random.normal(loc=18.48, scale=10.86, size=8000).clip(0, 50),
    'Review_Rating': np.random.normal(loc=4.14, scale=0.94, size=8000).clip(1, 5)
}

df = pd.DataFrame(data)

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Histograms
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# 2. Box Plots
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[column])
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()

# 3. Scatter Plots
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Customer_Age', y='Price')
plt.title('Price vs. Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Review_Rating', y='Price')
plt.title('Price vs. Review Rating')
plt.xlabel('Review Rating')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Purchase_History', y='Review_Rating')
plt.title('Purchase History vs. Review Rating')
plt.xlabel('Purchase History')
plt.ylabel('Review Rating')
plt.show()

# 4. Pair Plot
sns.pairplot(df)
plt.suptitle('Pair Plot of All Variables', y=1.02)
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
